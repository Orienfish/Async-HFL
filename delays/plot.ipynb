```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df1 = pd.read_csv('delay_client_to_gateway.csv', sep=',', header=None)

# Drop the rows that are all zeros and write back
#df = df.loc[(df != 0).any(1)]
#df.to_csv('delays/new.csv', sep=',', header=False, index=False)

num = 184

df1 = df1.replace(0, np.nan)
print(df1.shape)
print(df1.iloc[:num].shape)
print(df1.iloc[:num].min(axis=1).shape)
x_pos = np.arange(num)
data = df1.iloc[:num].min(axis=1)
fig, ax = plt.subplots(figsize=(4,3))
ax.bar(x_pos, data.sort_values(), align='center', 
       color='#56B4E9', ecolor='black', capsize=10)
ax.set_ylabel('Round Delay (s)')
ax.set_xlabel('RPi ID')

df2 = pd.read_csv('comp.txt', sep=',', header=None)
print(df2.shape)
print(df2.iloc[:num].min(axis=1).shape)

x_pos = np.arange(num)
data = df1.iloc[:num].min(axis=1)+df2.iloc[:num].min(axis=1)
print(data.sort_values())
# Build the plot
fig, ax = plt.subplots(figsize=(4.5,3.5))
ax.bar(x_pos, data.sort_values() / 1000, align='center', 
       color='#56B4E9', ecolor='black', capsize=10)
ax.set_ylabel('Round Delay (K s)', fontsize=14)
ax.set_xlabel('Edge Device ID', fontsize=14)
#ax.set_xticks(x_pos)
plt.xticks(fontsize=14)
plt.yticks([0, 0.5, 1, 1.5, 2], fontsize=14)
ax.yaxis.grid(True)
ax.set_ylim(0, 2)
ax.set_xlim(0, 183)
plt.tight_layout()
plt.savefig('delays.png')

fig, ax = plt.subplots(figsize=(4,3))
ax.hist(data, align='mid')
ax.set_ylabel('Round Delay (s)')
ax.set_xlabel('RPi ID')
ax.yaxis.grid(True)
```
